```python
#imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load Data
df = pd.read_csv("SPRAS_output\data0-pathway-summary.txt", sep = '\t')
```

```python
#Inspect
df.head()
```

```python
# Plot a histogram for the "Number of nodes" column using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['Nodes in prize'], bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Nodes in prize')
plt.xlabel('Nodes in prize')
plt.ylabel('Frequency')
plt.show()

# Alternatively, using Seaborn for a more styled plot
plt.figure(figsize=(10, 6))
sns.histplot(df['Number of nodes'], bins=30, color='blue')
plt.title('Histogram of Number of Nodes')
plt.xlabel('Number of Nodes')
plt.ylabel('Frequency')
plt.show()
```
