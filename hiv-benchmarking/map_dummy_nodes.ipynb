```python
human_nodes['uniprot_kb_protein'] = human_nodes[0] + "_" + human_nodes[1]
human_nodes
```

```python
human_nodes.to_csv("hiv_processed_data/human_dummy_proteins_uniprotkb.txt", index=False,columns=['uniprot_kb_protein'], header=False)
```
